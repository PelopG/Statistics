```python
import numpy as np # import numpy library

np.random.seed(0) # same random numbers appearing - predictable

n = 1000 # size 

# Procedure of X1 is the same as X2
# creating a numpy array starting with probability 0 - 1 of size n [np.random.uniform(low=0, high=1, size=n) --> condition ]
# the condition generated should be less than 0.3 
# Using numpy.where with AND logic return true if < 0.3 set 1 else set 0.
X1 = np.where(np.random.uniform(low=0, high=1, size=n) < 0.3, 1, 0) 
X2 = np.where(np.random.uniform(low=0, high=1, size=n) < 0.03, 1, 0) 
```

```python
import matplotlib.pyplot as plt # importing matplot library
%matplotlib inline

nn = np.arange(1, n + 1) # creating an evenly space array start from 1 to 1001

plt.figure(figsize=(12, 8)) # Create a new figure / Width, height in inches
plt.plot(nn, np.cumsum(X1) / nn, label='p = 0.3') # Return the cumulative sum of the elements along a given axis.
plt.plot(nn, np.cumsum(X2) / nn, label='p = 0.03')
plt.legend(loc='upper right')
plt.show()
```
